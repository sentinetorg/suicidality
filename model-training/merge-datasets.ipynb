# Merge all cleaned datasets

```python
SOURCE_DIR = "datasets/source"
TARGET_DIR = "datasets/extracted"
MERGED_FILE_NAME = "dataset-cleaned.csv"
```

```python
import glob, os
from zipfile import ZipFile
import pandas as pd

os.makedirs(TARGET_DIR, exist_ok=True)
```

```python
for file in glob.glob(f'{SOURCE_DIR}/*.zip'):
    with ZipFile(file, 'r') as zip:
        print(f'EXTRACTING: {file}', end=" ")
        zip.extractall(path=TARGET_DIR)
        print("[DONE]")
```

```python
merged_df = pd.DataFrame()

for file in glob.glob(f'{TARGET_DIR}/batches/*.csv'):
    try:
        df = pd.read_csv(file, lineterminator='\n')
        merged_df = pd.concat([merged_df, df], axis=0, ignore_index=True)
        
    except Exception as e:
        print(f'ERROR: {str(e)}')
        break

merged_df = merged_df.rename(columns={'Unnamed: 0': 'idx'})

merged_df
```

```python
merged_df.to_csv(f'{TARGET_DIR}/{MERGED_FILE_NAME}')
```
